```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms,models

import json,cv2
import numpy as np
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
# dataset
class KeypointsDataset(Dataset):
    def __init__(self, img_dir, data_file):
       self.img_dir = img_dir
       with open(data_file,"r")as f:
        self.data = json.load(f)

       self.transform = transforms.Compose([
           transforms.ToPILImage(),
           transforms.Resize((224,224)),
           transforms.ToTensor(),
           transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
       ])

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
      item = self.data[index]
      img_path = f"{self.img_dir}/{item['id']}.png"
      img = cv2.imread(img_path)
      h,w = img.shape[:2]

      img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      img = self.transform(img)

      kps = np.array(item['kps']).flatten()
      kps = kps.astype(np.float32)

      kps[::2] *= 224/w         # fixing x coordinates for resized image
      kps[1::2] *= 224/h        # fixing y coordinates for resized image

      return img, kps
```

```python
train_dataset = KeypointsDataset(
    img_dir = "data/images",
    data_file = "data/data_train.json"
)
val_dataset = KeypointsDataset(
    img_dir = "data/images",
    data_file = "data/data_val.json"
)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
# Model creation
model = models.resnet50(pretrained=True)
model.fc = torch.nn.Linear(model.fc.in_features, 14*2) # replace last layer to get keypoints
model = model.to(device)
```

```python
# Model training
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
epochs = 20
for epoch in range(epochs):
  for i,(images,kps) in enumerate(train_loader):
    images = images.to(device)
    kps = kps.to(device)

    optimizer.zero_grad()
    outputs = model(images)
    loss = criterion(outputs, kps)

    loss.backward()
    optimizer.step()

    if i % 10 == 0:
      print(f"Epoch: {epoch+1}/{epochs}, Batch: {i+1}/{len(train_loader)}, Loss: {loss.item()}")
```

```python
torch.save(model.state_dict(), "keypoints_model.pth")
```
